```python
# create corpus df.
product_corpus_df = df_verified.select("asin", "title", "text").dropna()
```

```python
# concat title, text as input 
from pyspark.sql.functions import concat_ws
product_corpus_df = product_corpus_df.withColumn("content", concat_ws(". ", "title", "text"))
```

```python
# get unique product info: asin, content 
unique_products = product_corpus_df.dropDuplicates(["asin"]).select("asin", "content")
```

```python
local_products = unique_products.limit(1000).toPandas()  # 先测试 1000 条
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")  # 可换成你喜欢的模型
embeddings = model.encode(local_products["content"].tolist(), show_progress_bar=True)
```

```python
import faiss
import numpy as np

index = faiss.IndexFlatL2(embeddings.shape[1])
index.add(np.array(embeddings))

# 模拟一个用户喜好向量，进行 Top-N 检索
D, I = index.search(np.array([embeddings[0]]), k=5)
recommended_items = local_products.iloc[I[0]]
```

```python
from transformers import AutoTokenizer, AutoModel
import torch
import numpy as np

# 模型名可换为 BAAI/bge-large-en
model_name = "BAAI/bge-small-en"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
model.eval()

def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output[0]  # [batch_size, tokens, embedding_dim]
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size())
    return (token_embeddings * input_mask_expanded).sum(1) / input_mask_expanded.sum(1)

def get_embedding(texts):
    with torch.no_grad():
        encoded_input = tokenizer(texts, padding=True, truncation=True, return_tensors="pt", max_length=512)
        model_output = model(**encoded_input)
        return mean_pooling(model_output, encoded_input['attention_mask']).cpu().numpy()

# 例如对 local_products["content"] 生成向量：
texts = local_products["content"].tolist()
embeddings = get_embedding(texts)
```
