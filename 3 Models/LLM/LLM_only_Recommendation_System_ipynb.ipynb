# Step- by -Step explanation of each step of pipeline

## Step 1 : Extract reviews


````python

def extract_user_reviews(df_reviews, df_metadata, user_ids):
    results = []
    for user_id in user_ids:
        user_reviews = df_reviews[df_reviews["user_id"] == user_id]
        reviews_with_titles = []
        for _, row in user_reviews.iterrows():
            asin = row["asin"]
            title = df_metadata[df_metadata["parent_asin"] == asin]["title"].values
            product_title = title[0] if len(title) > 0 else "Unknown Product"
            reviews_with_titles.append(f"{product_title}: {row['text']}")
        results.append(reviews_with_titles)
    return results
````

## Step 2 (LLM Summarisation of User Preference)

````python

# Summarization function
def summarize_user_preferences_direct(review_lists, max_new_tokens=400):
    summaries = []

    for review_list in review_lists:
        if not review_list:
            summaries.append("Not enough reviews to summarize preferences.")
            continue


        prompt = (
            "You are a consumer behavior analyst.\n"
            "Below are product reviews written by a single customer. Each review is about a specific product.\n\n"
            "Think carefully about these reviews and summarize the customer's shopping behavior in exactly 3 sentences:\n"
            "1. What are the common traits or issues among the products they reviewed?\n"
            "2. What seems to matter most to this customer when buying a product?\n"
            "3. What types of products is this customer likely to purchase in the future?\n\n"
            "Be concise, non-repetitive, and analytical.\n\n"
            "Reviews:\n" +
            "\n".join(f"- {r}" for r in review_list) +
            "\n\nSummary:"
        )

        # Tokenize
        input_ids = tokenizer(prompt, return_tensors="pt", truncation=True).input_ids.to(model.device)

        # Generate
        with torch.no_grad():
            outputs = model.generate(
                input_ids=input_ids,
                max_new_tokens=max_new_tokens,
                do_sample=False,
                temperature=0.0
            )

        # Decode
        output_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

        # Extract only the summary
        summary = output_text.split("Summary:")[-1].strip()
        summaries.append(summary)

    return summaries
````

## STEP 3: Filter by ratings and # of reviews

````python
def filter_items_by_rating(df_metadata, min_rating=4.0, min_review_count=5):
    """
    Filters items in the metadata dataframe by minimum average rating and review count.

    Parameters:
    - df_metadata (pd.DataFrame): DataFrame with at least 'average_rating' and 'rating_number' columns
    - min_rating (float): Minimum acceptable average rating
    - min_review_count (int): Minimum number of reviews required

    Returns:
    - pd.DataFrame: Filtered DataFrame with only 'title' and 'features' columns
    """
    return df_metadata[
        (df_metadata['average_rating'] >= min_rating) &
        (df_metadata['rating_number'] >= min_review_count)
    ][['title', 'features']].copy()

filtered_metadata = filter_items_by_rating(df_metadata)
````

## STEP 4:  Making top 300 using title and features

````python

#torch.tensor([1.0]).cpu().numpy()

def match_top_items_by_title(user_summary, filtered_metadata, top_k=300, model_name="all-MiniLM-L6-v2"):
    """
    Match top-k items to a user's preference summary using SBERT embeddings of product titles.

    Parameters:
    - user_summary (str): Summary text describing the user's preferences
    - filtered_metadata (pd.DataFrame): Product metadata (must contain 'title' column)
    - top_k (int): Number of top similar items to return
    - model_name (str): Name of the SentenceTransformer model

    Returns:
    - pd.DataFrame: Top-k most similar items based on cosine similarity
    """
    # Load SBERT model
    model = SentenceTransformer(model_name, device='cpu')

    # Encode item titles and user summary
    item_titles = filtered_metadata["title"].fillna("").tolist()
    item_embeddings = model.encode(item_titles, convert_to_tensor=True)
    user_embedding = model.encode([user_summary], convert_to_tensor=True)[0]

    item_embeddings = item_embeddings.cpu().numpy()
    user_embedding = user_embedding.cpu().numpy()

    # Normalize vectors for cosine similarity
    faiss.normalize_L2(item_embeddings)
    faiss.normalize_L2(np.expand_dims(user_embedding, axis=0))

    # Build FAISS index and search for top-k similar items
    index = faiss.IndexFlatIP(item_embeddings.shape[1])
    index.add(item_embeddings)
    _, indices = index.search(np.expand_dims(user_embedding, axis=0), top_k)

    # Retrieve top-k matched items
    top_items = filtered_metadata.iloc[indices[0]].copy()

    # Free memory
    del index, model
    torch.cuda.empty_cache()

    return top_items


````

## Step 5: Making top 50 using title and features

````python

def match_top_items_by_title_and_features(user_summary, metadata_df, top_k=50, model_name="all-MiniLM-L6-v2"):
    """
    Match top-k items using SBERT by combining product title and features.

    Parameters:
    - user_summary (str): Summary text describing the user's preferences.
    - metadata_df (pd.DataFrame): Product metadata (must contain 'title' and 'features').
    - top_k (int): Number of top similar items to return.
    - model_name (str): Name of the SentenceTransformer model.

    Returns:
    - pd.DataFrame: Top-k most similar items based on cosine similarity.
    """
    model = SentenceTransformer(model_name, device='cpu')

    product_texts = (metadata_df["title"].fillna("") + " " + metadata_df["features"].fillna("").astype(str)).tolist()

    item_embeddings = model.encode(product_texts, convert_to_numpy=True)
    user_embedding = model.encode([user_summary], convert_to_numpy=True)[0]

    faiss.normalize_L2(item_embeddings)
    faiss.normalize_L2(np.expand_dims(user_embedding, axis=0))

    index = faiss.IndexFlatIP(item_embeddings.shape[1])
    index.add(item_embeddings)
    _, indices = index.search(np.expand_dims(user_embedding, axis=0), top_k)

    top_items = metadata_df.iloc[indices[0]].copy()

    return top_items
````

## Last Step: Using LLM to generate top 5 with reason

### 🧠 Prompt Engineering Summary for Product Ranking (with LLM)

Generate a structured ranking of 5 products that best match a user's shopping preferences using a powerful LLM (e.g., Mistral 7B).

---

### ✅ Techniques Applied

#### 1. **Role Prompting**
Assigns a clear role to the model:
> *"You are a consumer behavior analyst."*

#### 2. **Structured Prompting**
Controls output format with explicit instruction:
> *Respond in the following format: Rank, Product, Keywords, Reason*

#### 3. **Deliberation Prompting**
Encourages reasoning before answering:
> *"Think carefully about the reviews..."*

#### 4. **Keyword Extraction Instead of Scoring**
Replaces numeric scores with semantically meaningful keywords to reduce hallucination and improve interpretability.

#### 5. **Example Priming (Few-shot Prompting)**
Provides a clear sample to guide the model:
```
Rank 1: Product: Coconut Hair Milk  
Matched Keywords: vegan, moisturizing, coconut oil  
Reason: The product aligns with the user's preference for natural ingredients and hydration-focused hair care.  
---
```

#### 6. **Output Constraint**
Constrains generation to avoid over/under-generation:
> *"(Repeat this format for exactly 5 products)"*

---

```text
Please respond in the following format:
Rank 1: Product: [Product Title]  
Matched Keywords: [keyword1, keyword2, keyword3]  
Reason: [Short reason]  
---
(Repeat this format for exactly 5 products)
```


````python
def rank_top_items_with_mistral(user_summary, top_items, max_new_tokens=800):
    """
    Use globally loaded tokenizer/model to rank top 5 items based on user summary and product descriptions.
    Prompts are formatted with explicit instructions and examples to guide the model output.
    """

    # Prepare product descriptions
    product_texts = (
        top_items["title"].fillna("") + ". " +
        top_items["features"].fillna("").astype(str)
    ).tolist()

    # Prompt engineering with structure + instruction
    prompt = (
        "You are a consumer behavior analyst.\n"
        "Based on the following customer preferences and product descriptions, rank the top 5 products.\n\n"
        f"User Preferences:\n{user_summary}\n\n"
        "You MUST use the customer's summary as the MAIN decision factor.\n"
        "Do NOT select products that do not reflect the customer's values.\n"
        "Highlight relevant keywords from the customer summary when explaining why a product is a good match.\n\n"
        "Here are the product candidates:\n"
    )

    for idx, product_text in enumerate(product_texts):
        trimmed = product_text[:300]  # prevent overflow
        prompt += f"{idx+1}. {trimmed}\n"

    prompt += (
        "\nPlease respond in the following format:\n"
        "Rank 1: Product: [Product Title]\n"
        "Reason: [Short reason - mention user preferences]\n"
        "---\n"
        "(Repeat for exactly 5 products)\n"
        "Start of answer:\n"
    )

    # Tokenize with attention mask + pad_token_id
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, padding=True)
    input_ids = inputs["input_ids"].to(model.device)
    attention_mask = inputs["attention_mask"].to(model.device)

    # Generate output
    with torch.no_grad():
        outputs = model.generate(
            input_ids=input_ids,
            attention_mask=attention_mask,
            max_new_tokens=max_new_tokens,
            do_sample=False,
            pad_token_id=model.config.eos_token_id  # set explicitly
        )

    output_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Extract only answer part
    ranked_result = output_text.split("Start of answer:")[-1].strip()

    return ranked_result
````

````python
def rank_top_items_with_mistral_2(user_summary, top_items, max_new_tokens=800):
    """
    Use globally loaded tokenizer/model to rank top 5 items based on user summary and product descriptions.
    """

    # Prepare product descriptions
    product_texts = (
        top_items["title"].fillna("") + ". " +
        top_items["features"].fillna("").astype(str)
    ).tolist()

    # Prompt with answer tag + focused reason constraint
    prompt = (
        "You are a consumer behavior analyst.\n"
        "Based on the following customer preferences and product descriptions, rank the top 5 products.\n\n"
        f"User Preferences:\n{user_summary}\n\n"
        "Important Instructions:\n"
        "- You MUST rank 5 products based primarily on the customer summary above.\n"
        "- Do NOT include products that clearly contradict the user's preferences.\n"
        "- Your reason MUST directly reflect a key point in the user summary.\n"
        "- Be brief and clear. 1-2 sentences per reason.\n\n"
        "Here are the product candidates:\n"
    )

    for idx, product_text in enumerate(product_texts):
        trimmed = product_text[:300]
        prompt += f"{idx+1}. {trimmed}\n"

    # Add format instructions + START OF ANSWER
    prompt += (
        "\n---START OF ANSWER---\n"
        "Respond in the following format:\n"
        "Rank 1: Product: [Product Title]\n"
        "Reason: [1-2 sentence explanation tied to user summary]\n"
        "---\n"
        "(Repeat for Ranks 2–5)"
    )

    # Tokenize + generate using global model/tokenizer
    input_ids = tokenizer(prompt, return_tensors="pt", truncation=True).input_ids.to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            input_ids=input_ids,
            max_new_tokens=max_new_tokens,
            do_sample=False,
            temperature=0.0
        )

    output_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Extract only the ranked part after the answer tag
    ranked_result = output_text.split("---START OF ANSWER---")[-1].strip()

    return ranked_result
````

# Final Pipeline

Takes user id as an input, produce top 5 ranking for the user


````python
from sklearn.base import BaseEstimator, TransformerMixin

# Step 1: Extract user reviews

class UserReviewExtractor(BaseEstimator, TransformerMixin):
    def __init__(self, df_reviews, df_metadata):
        self.df_reviews = df_reviews
        self.df_metadata = df_metadata

    def fit(self, X, y=None):
        return self

    def transform(self, user_ids):
        results = []
        for user_id in user_ids:
            user_reviews = self.df_reviews[self.df_reviews["user_id"] == user_id]
            reviews_with_titles = []
            for _, row in user_reviews.iterrows():
                asin = row["asin"]
                title = self.df_metadata[self.df_metadata["parent_asin"] == asin]["title"].values
                product_title = title[0] if len(title) > 0 else "Unknown Product"
                reviews_with_titles.append(f"{product_title}: {row['text']}")
            results.append(reviews_with_titles)
        return results

# Step 2: Summarize user preferences

class LLMPreferenceSummarizer(BaseEstimator, TransformerMixin):
    def __init__(self, tokenizer, model, max_new_tokens=400):
        self.tokenizer = tokenizer
        self.model = model
        self.max_new_tokens = max_new_tokens

    def fit(self, X, y=None):
        return self

    def transform(self, review_lists):
        summaries = []

        for review_list in review_lists:
            if not review_list:
                summaries.append("Not enough reviews to summarize preferences.")
                continue

            prompt = (
                "You are a consumer behavior analyst.\n"
                "Below are product reviews written by a single customer. Each review is about a specific product.\n\n"
                "Think carefully about these reviews and summarize the customer's shopping behavior in exactly 3 sentences:\n"
                "1. What are the common traits or issues among the products they reviewed?\n"
                "2. What seems to matter most to this customer when buying a product?\n"
                "3. What types of products is this customer likely to purchase in the future?\n\n"
                "Be concise, non-repetitive, and analytical.\n\n"
                "Reviews:\n" +
                "\n".join(f"- {r}" for r in review_list) +
                "\n\nSummary:"
            )

            input_ids = self.tokenizer(prompt, return_tensors="pt", truncation=True).input_ids.to(self.model.device)

            with torch.no_grad():
                outputs = self.model.generate(
                    input_ids=input_ids,
                    max_new_tokens=self.max_new_tokens,
                    do_sample=False,
                    temperature=0.0
                )

            output_text = self.tokenizer.decode(outputs[0], skip_special_tokens=True)
            summary = output_text.split("Summary:")[-1].strip()
            summaries.append(summary)

        return summaries[0]

# Step 3: Filter by rating and review count
class ItemRatingFilter(BaseEstimator, TransformerMixin):
    def __init__(self, df_metadata, min_rating=4.0, min_review_count=5):
        self.df_metadata = df_metadata
        self.min_rating = min_rating
        self.min_review_count = min_review_count

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        filtered = self.df_metadata[
            (self.df_metadata['average_rating'] >= self.min_rating) &
            (self.df_metadata['rating_number'] >= self.min_review_count)
        ][['title', 'features']].copy()

        return {
            "user_summary": X,
            "filtered_metadata": filtered
        }

# Step 4: First SBERT Matcher (using title only)
class FirstSBERTMatcher(BaseEstimator, TransformerMixin):
    def __init__(self, top_k=300, model_name="all-MiniLM-L6-v2"):
        self.top_k = top_k
        self.model_name = model_name

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        user_summary = X["user_summary"]
        filtered_metadata = X["filtered_metadata"]

        model = SentenceTransformer(self.model_name, device='cpu')
        item_titles = filtered_metadata["title"].fillna("").tolist()
        item_embeddings = model.encode(item_titles, convert_to_tensor=True)
        user_embedding = model.encode([user_summary], convert_to_tensor=True)[0]

        item_embeddings = item_embeddings.cpu().numpy()
        user_embedding = user_embedding.cpu().numpy()

        faiss.normalize_L2(item_embeddings)
        faiss.normalize_L2(np.expand_dims(user_embedding, axis=0))

        index = faiss.IndexFlatIP(item_embeddings.shape[1])
        index.add(item_embeddings)
        _, indices = index.search(np.expand_dims(user_embedding, axis=0), self.top_k)

        top_items = filtered_metadata.iloc[indices[0]].copy()

        return {
            "user_summary": user_summary,
            "top300_metadata": top_items
        }

# Step 5: Second SBERT Matcher (using title + features)
class SecondSbertMatcher(BaseEstimator, TransformerMixin):
    def __init__(self, top_k=50, model_name="all-MiniLM-L6-v2"):
        self.top_k = top_k
        self.model_name = model_name

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        user_summary = X["user_summary"]
        top300_metadata = X["top300_metadata"]

        model = SentenceTransformer(self.model_name, device='cpu')
        product_texts = (top300_metadata["title"].fillna("") + " " + top300_metadata["features"].fillna("").astype(str)).tolist()

        item_embeddings = model.encode(product_texts, convert_to_numpy=True)
        user_embedding = model.encode([user_summary], convert_to_numpy=True)[0]

        faiss.normalize_L2(item_embeddings)
        faiss.normalize_L2(np.expand_dims(user_embedding, axis=0))

        index = faiss.IndexFlatIP(item_embeddings.shape[1])
        index.add(item_embeddings)
        _, indices = index.search(np.expand_dims(user_embedding, axis=0), self.top_k)

        top_items = top300_metadata.iloc[indices[0]].copy()

        return {
            "user_summary": user_summary,
            "top50_metadata": top_items
        }

# === Step 6: Final Reranker with Mistral ===
class LLMReranker(BaseEstimator, TransformerMixin):
    def __init__(self, tokenizer, model, max_new_tokens=800):
        self.tokenizer = tokenizer
        self.model = model
        self.max_new_tokens = max_new_tokens

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        user_summary = X["user_summary"]
        top_items = X["top50_metadata"]

        product_texts = (
            top_items["title"].fillna("") + ". " + top_items["features"].fillna("").astype(str)
            ).tolist()

        prompt = (
            "You are a consumer behavior analyst.\n"
            "Based on the following customer preferences and product descriptions, rank the top 5 products.\n\n"
            f"User Preferences:\n{user_summary}\n\n"
            "Important Instructions:\n"
            "- You MUST rank 5 products based primarily on the customer summary above.\n"
            "- Do NOT include products that clearly contradict the user's preferences.\n"
            "- Your reason MUST directly reflect a key point in the user summary.\n"
            "- Be brief and clear. 1–2 sentences per reason.\n\n"
            "Here are the product candidates:\n"
        )

        for idx, product_text in enumerate(product_texts):
            prompt += f"{idx+1}. {product_text[:300]}\n"

        prompt += (
            "\n---START OF ANSWER---\n"
            "Respond in the following format:\n"
            "Rank 1: Product: [Product Title]\n"
            "Reason: [1–2 sentence explanation tied to user summary]\n"
            "---\n"
            "(Repeat for Ranks 2–5)"
        )

        input_ids = self.tokenizer(prompt, return_tensors="pt", truncation=True, padding=True).input_ids.to(self.model.device)

        with torch.no_grad():
            outputs = self.model.generate(
                input_ids=input_ids,
                max_new_tokens=self.max_new_tokens,
                do_sample=False,
                temperature=0.0,
                pad_token_id=self.tokenizer.eos_token_id
            )

        output_text = self.tokenizer.decode(outputs[0], skip_special_tokens=True)
        ranked_result = output_text.split("---START OF ANSWER---")[-1].strip()

        return [ranked_result]


````

### Final Pipeline test!

````python
# === Setup Updated Pipeline ===
pipeline = Pipeline([
    ("extract_reviews", UserReviewExtractor(df_reviews, df_metadata)),
    ("summarize_user", LLMPreferenceSummarizer(tokenizer,model)),
    ("filter_items", ItemRatingFilter(df_metadata)),
    ("first_match", FirstSBERTMatcher()),
    ("second_match", SecondSbertMatcher()),
    ("llm_ranker", LLMReranker(tokenizer,model))
])
````

````python
import re

def extract_reasons_from_recommendation(text_block):
    """
    Extracts the 5 reason strings from the recommendation output.
    """
    matches = re.findall(r"Reason:\s*(.*?)(?:\n|$)", text_block)
    return matches[:5]  # Top 5만 사용
````

````python
from sentence_transformers import SentenceTransformer, util

def calculate_ad_dup_ratio(reasons, threshold=0.9):
    """
    Calculates the Ad Duplication Ratio among top-5 reasons.
    """
    model = SentenceTransformer("all-MiniLM-L6-v2")
    embeddings = model.encode(reasons, convert_to_tensor=True)

    duplicates = 0
    total_pairs = 0

    for i in range(len(reasons)):
        for j in range(i+1, len(reasons)):
            sim = util.cos_sim(embeddings[i], embeddings[j]).item()
            if sim >= threshold:
                duplicates += 1
            total_pairs += 1

    return duplicates / total_pairs if total_pairs > 0 else 0
def calculate_avg_cos_sim(user_summary, recommendations_texts):
    """
    Computes average cosine similarity between user summary and product + reason descriptions.
    """
    model = SentenceTransformer("all-MiniLM-L6-v2")
    user_emb = model.encode(user_summary, convert_to_tensor=True)

    similarities = []
    for text in recommendations_texts:
        product_desc = re.search(r"Product:\s*(.*?)\n", text)
        reason = re.search(r"Reason:\s*(.*?)\n", text)
        if product_desc and reason:
            combined = product_desc.group(1) + ". " + reason.group(1)
            item_emb = model.encode(combined, convert_to_tensor=True)
            sim = util.cos_sim(user_emb, item_emb).item()
            similarities.append(sim)

    return sum(similarities) / len(similarities) if similarities else 0
````
